## Working code


```python
mainurl="https://www.hltv.org"
url = "https://www.hltv.org/results?startDate=2019-08-06&endDate=2019-11-06"

headers = requests.utils.default_headers()
headers.update({
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',
})

page = requests.get(url, headers=headers)    
data = page.text
soup = BeautifulSoup(data)

array = []
for i in range(20):
    resultEL = soup.find_all("div", class_="result-con".split())

   
    for el in resultEL:
        game = el.get_text().strip().replace(" ", "").replace("\n", " ").split()
        if(game[4] == 'bo3'):
            a = el.find("a", href=True)

            if(a != None):
                gameUrl = mainurl + a['href']
                gamePage = requests.get(gameUrl, headers=headers) 
                gameData = gamePage.text
                gameSoup = BeautifulSoup(gameData)


                gameResult = gameSoup.find_all("div", class_="results")
                gameMap = gameSoup.find_all("div", class_="mapname")
                for x, y in zip(gameResult, gameMap):
                    score=x.get_text()[:5].replace(":", "-")
                    mapName=y.get_text()   
                    game = ([game[0], score, game[2], game[3] ,mapName])
                    array.append(game)
        array.append(game)
    if(i!=0):
        linkEL = soup.find_all("a", class_="pagination-next")
        nextPageURL = mainurl + linkEL[0]['href']
        page = requests.get(nextPageURL, headers=headers)  
        data = page.text
        soup = BeautifulSoup(data)

        
```

```python
results = pd.DataFrame(array)
results.columns = ['Team1', 'Score', 'Team2', 'Competition', 'Map']
results.to_csv("results.csv", index=False)
```
